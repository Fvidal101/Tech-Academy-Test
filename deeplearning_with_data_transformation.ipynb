```

# Creating the model for deep learning
from keras.models import Sequential
from keras.layers import Dense
import numpy as np 
import pandas as pd

```

```
data = pd.read_csv("ChurnData.csv")
data.head()
```

```
# change all the dataset into an array and the target variable into int types for binary values, i.e., churn
data = data[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip',   'callcard', 'wireless','churn']]
data['churn'] = data['churn'].astype('int')
data.head()
```

```
from sklearn import preprocessing

x = np.asarray(data[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip']])
y = np.asarray(data['churn'])

#We simply transform the data into more structures form; Mean = 0 
x = preprocessing.StandardScaler().fit(x).transform(x)
```

```
x
```

```

#Building the model 
model = Sequential()
model.add(Dense(24, input_dim=7, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```
# Fit the model
model.fit(x, y, epochs=500, batch_size=32)
```

```

# evaluate the model
scores = model.evaluate(x, y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
